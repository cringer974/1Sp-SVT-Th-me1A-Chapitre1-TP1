**Ressource 2** : création du graphique à partir de la Dataframe précédente.<br/>
**Etudier** et **exécuter** la cellule ci-dessous.<br/>

```python
# importation de la librairie matplotlib et son module pyplot spécialisé dans la création de graphique
import matplotlib.pyplot as plt 
import matplotlib.patches
from matplotlib.patches import Rectangle # les deux dernières lignes vont permettre de dessiner des rectangles

# stockage des valeurs des paramètres de la Dataframe dans les variables x et y, respectivement abscisse et ordonnée
x='Temps_(h)' 
y='Quantite_ADN_(UA)'

fig=plt.figure(figsize=(10,5)) # permet de créer un nouvelle figure mais instruction non obligatoire car automatique
ax = fig.add_subplot(111) # créé une sous figure à la première où seront créés des rectangles
rect1 = matplotlib.patches.Rectangle((0,0), 11, 1, color="y") # créé un rectangle avec des caractéristiques entre parenthèses
rect2 = matplotlib.patches.Rectangle((11,0), 1, 1, color="r")    # position du coin inférieur gauche (5,1), largeur (10) et hauteur (10) sur cette même ligne
rect3 = matplotlib.patches.Rectangle((12,0), 2, 1, color="y") 

plt.plot(mitose[x],mitose[y]) # utilise la dataframe "mitose" comme source de données
                              # pour créer un graphique 

# affichage d'une grille avec les lignes principales en gris foncé 
plt.grid(b=True, which='major', color='#666666', linestyle='--')       

# affiche les conditions du milieu (obscutité/lumière)
ax.add_patch(rect1)
ax.add_patch(rect2)
ax.add_patch(rect3)
plt.text(5,1.5,'Interphase')
plt.text(11.1,1.5,'Mitose')
plt.text(12.3,1.5,'Interphase')

# limites inférieures et supérieures des axes x et y
plt.xlim(0, 15)    # on définit les limites min et max de l'axe X
plt.ylim(0, 14)    # on définit les limites min et max de l'axe Y

# ajoute des titres au graphique
plt.title('Variation de la quantité ADN en fonction du temps') # donne le titre du graphique
plt.xlabel('Temps (heures)')  # donne le titre de l'axe X
plt.ylabel('Quantité ADN (UA)')  # donne le titre de l'axe Y

# affichage du graphique
plt.show() 

# Si on veut une sauvegarde du graphique sous la forme d'un fichier image plutôt que 
# que son affichage, on enlève le "#" à la ligne ci-dessous et on l'ajoute devant le "plt.show()" de la ligne 40
# plt.savefig('mitose_Jupyter.png'); plt.close('mitose_Jupyter.png')
```
