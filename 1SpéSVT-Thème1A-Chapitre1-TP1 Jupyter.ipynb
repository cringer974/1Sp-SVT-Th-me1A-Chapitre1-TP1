```python
# Importation les données de variation de la quantité d'ADN durant la mitose que vous avez déjà traitées
# sous excel et de créer une dataframe.

import pandas as pd # importe la librairie pandas et utilise l'alias pd

# création de la dataframe
mitose = pd.read_excel('1SpéSVT-Thème1A-Chapitre1-TP1.xlsx', sheet_name='Tab_mitose', skiprows=1, usecols=range(1,4)) 
# met le fichier Excel désigné dans une dataframe nommée mitose.
# on travaille sur l'onglet "Tab_mitose" du fichier "1SpéSVT-Thème1A-Chapitre1-TP1.xlsx". 
# la première ligne de la feuille Excel n'est pas prise en compte (d'où le skiprows)
# Seules les colonnes 1 (Temps_(h)) à 4 (Quantite_ADN_(UA)) sont utilisées. La colonne 0 du fichier étant vide, on l'élimine.

# affichage la dataframe
print(mitose) 

# La première colonne lisible dans la dataframe ci-dessous (chiffrée de 0 à 9) correspond à l'indexation des lignes de la dataframe et aide
# à localiser les cellules. La première ligne correspond à l'en-tête. ELle n'est pas comptabilisée et comporte les noms 
# des colonnes.
# Donc, le fonctionnement de la dataframe est comparable à celle d'une feuille de tableur.
```

```python
# importation de la librairie matplotlib et son module pyplot spécialisé dans la création de graphique
import matplotlib.pyplot as plt 

# désignation des éléments à prendre en compte pour le grahique
mitose.plot(kind='line',x='Temps_(h)',y='Quantite_ADN_(UA)') # utilise la dataframe "mitose" comme source de données
                                                         # pour le graphique en précisant que la colonne "temps" 
                                                         # est en abscisse et celle donnant "la quantité d'ADN" est en
                                                         # ordonnée

# affichage d'une grille avec les lignes principales en gris foncé 
plt.grid(b=True, which='major', color='#666666', linestyle='--')       

# limites inférieures et supérieures des axes x et y
plt.xlim(0, 15)    # on définit les limites min et max de l'axe X
plt.ylim(0, 14)    # on définit les limites min et max de l'axe Y

# ajoute des titres au graphique
plt.title('Variation de la quantité ADN en fonction du temps') # donne le titre du graphique
plt.xlabel('Temps (heures)')  # donne le titre de l'axe X
plt.ylabel('Quantité ADN (pg)')  # donne le titre de l'axe Y

# affichage du graphique
plt.show() 

# Si on veut une sauvegarde du graphique sous la forme d'un fichier image plutôt que 
# que son affichage, on enlève le "#" à la ligne ci-dessous et on l'ajoute devant le "plt.show " de la ligne 23
# plt.savefig('mitose_Jupyter.png'); plt.close('mitose_Jupyter.png')
```

9. <strong>Répondre</strong> au problème posé, dans la cellule ci-dessous, sous la forme d’un bilan donnant les différences entre les deux types de division en faisant référence à ce qui a été mis en évidence sur les graphiques, ce qui a été observé au microscope ainsi qu'aux docments fournis (/4). 
</p>

D’après la question 1, la méiose n’est pas une division conforme, contrairement à la méiose, du fait que la quantité d’ADN est divisée par 2.<br/> 
La question 2 montre que durant la métaphase de mitose les chromosomes à 2 chromatides se réunissent sur le plan équatorial de la cellule. Puis, en anaphase, les chromatides se séparent pour se diriger vers les pôles de la cellule.<br/>
La question 3 monte la suite du phénomène : durant la télophase, les le cytoplasme de la cellule-mère se scinde en 2 (cytodiérèse) et 2 nouveaux noyaux se forment pour donner naissance à 2 nouvelles cellules comportant les mêmes chromosomes que la cellule originale.<br/>
La question 4 permet de comprendre ce qui se déroule durant la méiose. Durant l’anaphase 1, les chromosomes à 2 chromatides d’une même paire se séparent. En télophase 1, chaque cellule-fille en formation reçoit la moitié des chromosomes de départ. Pendant la seconde partie de la méiose, les cellules-filles montrent une division de type mitose. A l’anaphase 2, les chromatides de chaque chromosome se séparent permettant, suite à la télophase, d’obtenir deux autres cellules-filles ayant la même quantité d’ADN. La première division de méiose est qualifiée de « réductionnelle », tandis que la seconde « d’équationnelle ». Au début de la méiose, la cellule-mère comporte des paires de chromosome. Elle est dite « diploïde ». La seconde division de méiose est dite « haploïde » : les cellules-filles ne gardent qu’un exemplaire de chaque paire de chromosome.
